# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data,x="Gender",palette="viridis")
plt.ylabel("Count")
plt.title('Gender Distribution')
ax.bar_label(ax.containers[(0)])
plt.show()
```

From The Given Graph The Number of females is greate than the number of Males

```python
gb =data.groupby('ParentEduc').agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
gb
```

```python
sns.heatmap(gb,annot=True,)
plt.title("Relationship Between Parents Education and Student's Score")
plt.figure(figsize=(4,4))
plt.show()
```

From the above chat that we concluded that eduction of parents play a great impact on student score
